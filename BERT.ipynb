```python

from transformers import BertTokenizer, BertModel

# Cargar el tokenizador y el modelo BERT preentrenado
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Texto de ejemplo para la tokenización y la representación
input_text = "BERT is a powerful model for natural language processing."

# Tokenizar el texto
tokens = tokenizer.tokenize(input_text)
print("Tokens:", tokens)

# Convertir tokens a IDs de vocabulario
input_ids = tokenizer.convert_tokens_to_ids(tokens)
print("Input IDs:", input_ids)

# Añadir padding a los input_ids
padding_length = 10 - len(input_ids)
padded_input_ids = input_ids + [0] * padding_length
print("Padded Input IDs:", padded_input_ids)

# Obtener la representación de tokens usando el modelo BERT
input_tensor = torch.tensor([padded_input_ids])
with torch.no_grad():
    output = model(input_tensor)

# Obtener la representación de la última capa oculta
last_hidden_states = output.last_hidden_state
print("Last Hidden States Shape:", last_hidden_states.shape)
```
